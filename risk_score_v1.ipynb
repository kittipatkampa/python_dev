# Calculate optimal $\alpha_1$ and $\alpha_2$

Calculate the performance (yield, coverage) at each threshold

```python
df['POST_30_GOODS_MFN'] = df['POST_30_ORDERS_MFN'] - df['POST_30_DEFECTS_MFN']
thr_orders = df[['POST_30_ORDERS_MFN','POST_30_GOODS_MFN','POST_30_DEFECTS_MFN']].groupby(factor).sum()
thr_records = df[['POST_30_ORDERS_MFN']].groupby(factor).size() ## number of records in each bin


## total 
total_orders = df['POST_30_ORDERS_MFN'].sum()
total_goods = df['POST_30_GOODS_MFN'].sum()
total_defects = df['POST_30_DEFECTS_MFN'].sum()
total_records = thr_records.sum() ## number of records

## Left
thr_orders['L_orders'] = thr_orders['POST_30_ORDERS_MFN'].cumsum()
thr_orders['L_goods'] = thr_orders['POST_30_GOODS_MFN'].cumsum()
thr_orders['L_defects'] = thr_orders['POST_30_DEFECTS_MFN'].cumsum()
thr_orders['perc_L_orders'] = 100*thr_orders['L_orders']/total_orders
thr_orders['perc_L_goods'] = 100*thr_orders['L_goods']/total_goods ## % goods coverage
thr_orders['perc_L_defects'] = 100*thr_orders['L_defects']/total_defects 
thr_orders['L_good_ratio'] = 100*thr_orders['L_goods']/thr_orders['L_orders'] ## purity of goods on L
thr_orders['L_records'] = thr_records.cumsum()
thr_orders['perc_L_records'] = 100*thr_orders['L_records']/total_records


## Right
thr_orders['R_orders'] = total_orders - thr_orders['L_orders']
thr_orders['R_goods'] = total_goods - thr_orders['L_goods']
thr_orders['R_defects'] = total_defects - thr_orders['L_defects']
thr_orders['perc_R_orders'] = 100*thr_orders['R_orders']/total_orders
thr_orders['perc_R_goods'] = 100*thr_orders['R_goods']/total_goods
thr_orders['perc_R_defects'] = 100*thr_orders['R_defects']/total_defects ## % defects coverage
thr_orders['R_defect_ratio'] = 100*thr_orders['R_defects']/thr_orders['R_orders'] ## purity of defects on R
thr_orders['R_records'] = total_records - thr_orders['L_records']
thr_orders['perc_R_records'] = 100*thr_orders['R_records']/total_records

thr_orders_records = thr_orders[['L_good_ratio','perc_L_goods','L_goods','L_defects','L_records','perc_L_records','R_defect_ratio','perc_R_defects','R_goods','R_defects','R_records','perc_R_records']]
thr_orders_records
```

# A better way to do quantile binning

It is not safe to use qcut on some data array. Below is an example where qcut works only for n < 5. That is because when n >= 5, the bin edges/separators are not unique, hence qcut gives error message:

ValueError: Bin edges must be unique: array([-99., -99.,  79.,  82.,  84.,  99.])

```python
n = 5
pd.qcut(X2,n)
```

Therefore, it's much safer to do the following.

```python
## first find the valid quantile edges, then dedup them
n = 10
bin_sep = X2.quantile(np.linspace(0.0, 1.0, num=n))
print('bin_sep w/o dedup\n',bin_sep)
bin_sep = bin_sep.unique()
print('bin_sep after deduped\n',bin_sep)

## Now you can use the bin_sep to factor the data
factor, bin_edges = pd.cut(x=X2, bins=bin_sep, right=True, include_lowest=True, retbins=True)
print('factor',factor)
print('bin_edges',bin_edges)
pd.DataFrame({'X2':X2, 'factor':factor})
```

# Calculate Weight of Evidence -- Spearman correlation

Start with a large number of intervals, keep reducing the number of interval until the Spearman correlation between X.mean and Y.mean is 1. 

Example here is adapted from 

https://statcompute.wordpress.com/2012/12/08/monotonic-binning-with-python/

```python
import scipy.stats.stats as stats
n = 100
r = 0
while np.abs(r) < 1:
    bin_sep = X2.quantile(np.linspace(0.0, 1.0, num=n))
    bin_sep = bin_sep.unique()
    factor,bins = pd.cut(x=X2, bins=bin_sep, right=True, include_lowest=True, retbins=True, precision=5)
    d1 = pd.DataFrame({"X": X2, "Y": Y, "Bucket": factor })
    d2 = d1.groupby('Bucket', as_index = True)
    r, p = stats.spearmanr(d2.mean().X, d2.mean().Y)
    print('n=',n,' r=',r)
    n = n-1

## Bucket, #neg, #pos, ratio_bad_ovr_good
counts = d2.size()
num_pos = d2.sum().Y
num_neg = d2.size()-num_pos
ratio_pos_ovr_neg = num_pos/num_neg
total_pos = sum(num_pos)
total_neg = sum(num_neg)
rel_p_pos = num_pos/total_pos
rel_p_neg = num_neg/total_neg
WoE = 100*np.log(rel_p_pos/rel_p_neg)

## report the results
print('final bins=',bins)
WoE_tbl = pd.DataFrame({'counts':counts, 'num_pos':num_pos, 'num_neg':num_neg, 'ratio_pos_ovr_neg':ratio_pos_ovr_neg, 'rel_p_pos':rel_p_pos, 'rel_p_neg':rel_p_neg, 'WoE':WoE})
pd.DataFrame(data=WoE_tbl, columns=['counts','num_pos','num_neg','ratio_pos_ovr_neg','rel_p_pos','rel_p_neg','WoE']) 
```

# Merging intervals using Chi-square test

$H_0$ -- A and B are independent P(A,B) = P(A)P(B)

$H_a$ -- A and B are not independent P(A,B) $\ne$ P(A)P(B)
      

If p-value >= 0.05 --> var1 and var2 are independent, hence f1 and f2 are from the same distribution.
If p-value < 0.05 --> var1 and var2 are NOT independent, hence f1 and f2 are from different distributions.
Note: f1 and f2 are treated as value instances for var2. Chi$^2$ test the independence of var1 and var2, not f1 and f2.

Example#1: The result shows p-value = 1.0, which is greater than 0.05. Therefore, we fail to reject the null hypothesis, hence two variables (var1 and var2) are independent and so f1 and f2 are said to come from the same distribution. 


```python
import scipy.stats as sp
f1 = np.array([80,20])
f2 = np.array([8,2])
conting = np.array([f1,f2])
##             var1
##            0    1
##      --------------
## var2 f1 | 80   20
##      f2 |  8    2
##      --------------
out = sp.chi2_contingency(conting)
out[0] ## chi2 test statistics
out[1] ## p-value of the test
out[2] ## Degrees of freedom
out[3] ## expected frequencies
print('chi2 statistics',out[0],' p-value',out[1],' degree of freedom',out[2],' expected freq',out[3])
```

Example#2 The p-value = 0.02 which is less than 0.05, therefore, we will reject the null hypothesis, two variables are not indepndent, hence f1 and f2 are not from the same distribution.

```python
f1 = np.array([80,20])
f2 = np.array([8,8])
conting = np.array([f1,f2])
out = sp.chi2_contingency(conting)
print('chi2 statistics',out[0],' p-value',out[1],' degree of freedom',out[2],' expected freq',out[3])
```

Example#3 It's more forgivable when having more data points. When using 800,200, the range within p-value=0.05 is wider than 80,20.

```python
p1 = 80
p2 = 20
epsilon_list = np.arange(-20,20,1)

def cal_p_val(p1,p2,epsilon_list):
    f1 = np.array([p1,p2])
    p_value = np.zeros_like(epsilon_list)*1.
    for i in np.arange(len(epsilon_list)):
        epsilon = epsilon_list[i]
        f2 = np.array([p1,p2+epsilon])
        #print(f2)
        conting = np.array([f1,f2])
        out = sp.chi2_contingency(conting)
        #print(out[1])
        p_value[i] = out[1]
        #print('i=',i,'  epsilon=',epsilon,'  p-value=',p_value[i])
    return (p_value)

epsilon_list1 = np.arange(-20,20,1)
p_value1 = cal_p_val(p1=80,p2=20,epsilon_list=epsilon_list1)
epsilon_list2 = np.arange(-200,200,1)
p_value2 = cal_p_val(p1=800,p2=200,epsilon_list=epsilon_list2)
plt.plot(epsilon_list1,p_value1, 'k-', label='80, 20')        
plt.plot(epsilon_list2,p_value2, 'r-', label='800,200')

```

# Merging bin intervals using Chi$^2$ test

```python
## Create dummy data
X = np.random.randint(0,100,10000)*0.00000001
X[np.random.randint(0,10000,100)] = None
X = pd.Series(data=X)
print(type(X))

Y = np.random.randint(0,2,10000)
Y = pd.Series(data=Y)
print(type(Y))

X_new = np.random.randint(-150,150,100)*0.00000001
X_new[np.random.randint(0,100,10)] = None
X_new = pd.Series(data=X_new)
print(type(X_new))

X[0:1000]
```

```python
## initial number of bins
n = 20 
## p-value threshold, if p-value is greater than this value, will merge
## the larger p_val_thr, the harder the intervals merge
p_val_thr = 0.7 #0.05 

## Get records that are not NULL
X1 = X[pd.notnull(X)]
Y1 = Y[pd.notnull(X)]
print('0/1 counts from not-null\n',Y1.value_counts())


## number of positive, negative records
total_counts = len(Y)
total_pos = sum(Y)
total_neg = total_counts-total_pos

## =======================================================================
## This part process only not-null records X1,Y1
## =======================================================================

## create bin edges
bin_sep = X1.quantile(np.linspace(0.0, 1.0, num=n))
bin_sep = bin_sep.unique()
print('initial bin_sep:\n',bin_sep)

done = 0 ## indicator
while done == 0:
    factor,bins = pd.cut(x=X1, bins=bin_sep, right=True, include_lowest=True, retbins=True, precision=5)
    d1 = pd.DataFrame({"X": X1, "Y": Y1, "Bucket": factor })
    d2 = d1.groupby('Bucket', as_index = True)

    ## Bucket, #neg, #pos, ratio_bad_ovr_good
    counts = d2.size()
    num_pos = d2.sum().Y
    num_neg = counts-num_pos
    ratio_pos_ovr_neg = num_pos/num_neg
    rel_p_pos = num_pos/total_pos
    rel_p_neg = num_neg/total_neg
    WoE = np.log(rel_p_pos/rel_p_neg)
    bin_edge_left = bins[:(len(bins)-1)]
    bin_edge_right = bins[1:len(bins)]
    min_intv = d2.min().X
    max_intv = d2.max().X

    ## output the results
    #print('final bins=',bins)
    profile_dict = pd.DataFrame({'counts':counts, 'num_pos':num_pos, 'num_neg':num_neg, 'ratio_pos_ovr_neg':ratio_pos_ovr_neg, 'rel_p_pos':rel_p_pos, 'rel_p_neg':rel_p_neg, 'WoE':WoE, 'min_intv':min_intv,'max_intv':max_intv})
    profile_tbl = pd.DataFrame(data=profile_dict, columns=['min_intv','max_intv','counts','num_pos','num_neg','ratio_pos_ovr_neg','rel_p_pos','rel_p_neg','WoE']) 
    ## sort the interval by min_value
    profile_tbl = profile_tbl.sort_index(by=['min_intv'])
    ## After sort, we can add the bin edge
    profile_tbl['bin_edge_left'] = bin_edge_left
    profile_tbl['bin_edge_right'] = bin_edge_right

    ## add auxiliary columns i, j and p-value
    profile_tbl['i'] = np.zeros(len(profile_tbl))
    profile_tbl['j'] = np.zeros(len(profile_tbl))
    profile_tbl['adjc_p_value'] = np.zeros(len(profile_tbl))

    ## ch^2-test and calculate p-value for each adjacent row pair
    i = 0
    while i < (len(profile_tbl)-1):
        #print(i)
        conting = profile_tbl[['num_pos','num_neg']].iloc[[i,i+1]]
        profile_tbl.adjc_p_value[i] = sp.chi2_contingency(conting)[1]
        profile_tbl.i[i] = i
        profile_tbl.j[i] = i+1
        #print(i,',',i+1,' p-value:',profile_tbl.adjc_p_value[i])
        i = i + 1

    print(profile_tbl)    
    ## Get the max p-value with corresponding index
    idx_p_val_max = profile_tbl.adjc_p_value.idxmax()
    p_val_max = profile_tbl.adjc_p_value[idx_p_val_max] ## the p-value max
    idx_to_remove = profile_tbl.j[idx_p_val_max] ## the index we want to remove in bins
    
    ## Compare the max p-value with the threshold
    if p_val_max >= p_val_thr:
        ## merge the interval i and j
        ## This is a new bin separators
        bin_sep = np.delete(bins,idx_to_remove)
        print('p_val_max=',p_val_max,'>= p_val_thr=',p_val_thr,'  continue merging intervales')
        done = 0
    else:
        ## all intervals are sufficiently unique, no more mergeing, done!!!  
        print('p_val_max=',p_val_max,'< p_val_thr=',p_val_thr,'  done merging intervales!!!')
        done = 1
    
    
    print('current bin separators',bins)
    print('new bin separators',bin_sep)

## final outputs for not-null records
print('\n')
print('-'*99,'\n')
print('final profile table: \n',profile_tbl)  
print('\n')
print('-'*99,'\n')
print('final bin: \n',bin_sep)

## more compact output
WoE_numer = profile_tbl[['bin_edge_left','bin_edge_right','WoE']]
```
