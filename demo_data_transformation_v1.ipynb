# pandas and scikit-learn

```python
from sklearn import tree
import pandas as pd
import numpy as np
```

First, generate the data points consisting of integer, float and categorical.

```python
## This is a function to generate data
def genData1(a1,a2,b1,b2,c,l,g,cnt):
    a=[randint(a1,a2) for i in range(cnt)]
    b=[b1*rand()+b2 for i in range(cnt)]
    c=[c for i in range(cnt)]
    label=[l for i in range(cnt)]
    group=[g for i in range(cnt)]
    df = pd.DataFrame({'a' : a, 'b' : b, 'c' : c,'label':label, 'group':group})
    return df

## Test data
group1 = genData1(0,10,0.5,0.5,'A',0,1,100)
group2 = genData1(0,5,0.5,0,'A',1,2,50)
group3 = genData1(5,10,0.5,0,'A',1,3,50)
group4 = genData1(0,5,0.5,0.5,'B',1,4,50)
group5 = genData1(5,10,0.5,0.5,'B',1,5,50)
group6 = genData1(0,10,0.5,0,'B',0,6,100)
df_train = pd.concat([group1, group2, group3, group4, group5, group6], ignore_index=True)
print('train size:',df_train.shape)

## Test data
group1 = genData1(0,10,0.5,0.5,'A',0,1,10)
group2 = genData1(0,5,0.5,0,'A',1,2,10)
group3 = genData1(5,10,0.5,0,'A',1,3,10)
group4 = genData1(0,5,0.5,0.5,'B',1,4,10)
group5 = genData1(5,10,0.5,0.5,'B',1,5,10)
group6 = genData1(0,10,0.5,0,'B',0,6,10)
df_test = pd.concat([group1, group2, group3, group4, group5, group6], ignore_index=True)
print('test size:',df_test.shape)
```

write the datasets to csv file for further use.

```python
df_train.to_csv(path_or_buf='df_train_tree.csv', sep=',', na_rep='', header=True, index=False)
df_test.to_csv(path_or_buf='df_test_tree.csv', sep=',', na_rep='', header=True, index=False)
```

## Using DictVectorizer

```python
from sklearn.feature_extraction import DictVectorizer
```

Convert from dataframe to `list of dictionary`. Note that the dictionary will not work for scikit-learn.

```python
## This will not work for scikit-learn. Don't do this.
#dict_train = df_train.to_dict()
#dict_train

## Instead do it. This is how I convert dataframe to list of dict.
## I believe there should be a more efficient way to this than using for-loop.
list_train=[]
for i in range(df_train.shape[0]):
    list_train.append(df_train[['a','b','c']].T.to_dict()[i])
print(list_train[:10])

y_train = array(df_train.label)
y_train
```

```python
vec = DictVectorizer()
vec
```

```python
train = vec.fit_transform(list_train).toarray()
train
```

```python
## This is the resulting feature names. 
## Note that the numerical variables are not modified. 
## Only the categorical (e.g. string) variables are 'binarized'.

vec.feature_names_
```

```python
## Convert the test data 
list_test=[]
for i in range(df_test.shape[0]):
    list_test.append(df_test[['a','b','c']].T.to_dict()[i])
print(list_test[:10])

y_test = array(df_test.label)
print(y_test)

## Note that we use the transformation "vec" from the previous step.
## In the previous step, we use vec.fit_transform() to build the transformation vec from the training data.
## So, this time we will use vec.transform() to transform the test dataset.
test = vec.transform(list_test).toarray()
test[:10]
```

Finally, we will build a decision tree from scikit-learn using `train`.

```python
clf = tree.DecisionTreeClassifier(criterion='gini', splitter='best', min_samples_split=20)
clf = clf.fit(train, y_train)
```

## Plot the decision tree

And we will plot the learned decision tree

```python
from sklearn.externals.six import StringIO  

## This package will need package pydot which only works for python2.7
import pydot 

## Extract the decision tree logic from the trained model
dot_data = StringIO() 
tree.export_graphviz(clf, out_file=dot_data) 

## convert the logics into graph
graph = pydot.graph_from_dot_data(dot_data.getvalue()) 

## This will plot decision tree in pdf file
graph.write_pdf(path="test_decistion_tree.pdf") 

from IPython.display import Image #needed to render in notebook
Image(graph.create_png())
```

## Plot decision surface of pair-feature

Now, we will plot the decision surface of the decision tree. I modify the original code from 
http://scikit-learn.org/stable/auto_examples/tree/plot_iris.html#example-tree-plot-iris-py


```python
## Now we will plot the decision surface
import matplotlib.pyplot as plt
import numpy as np

# Parameters
n_classes = 2
plot_colors = "bry"
plot_step = 0.02

# Load data
#iris = load_iris()
target_name = ['0','1']

for pairidx, pair in enumerate([[0, 1],[0,2],[0,3],[1, 2],[1,3],[2,3]]):
    # We only take the two corresponding features
    X = train[:, pair]
    y = y_train

    # Shuffle
    idx = np.arange(X.shape[0])
    np.random.seed(13)
    np.random.shuffle(idx)
    X = X[idx]
    y = y[idx]

    # Standardize
    #mean = X.mean(axis=0)
    #std = X.std(axis=0)
    #X = (X - mean) / std

    # Train
    clf = tree.DecisionTreeClassifier(criterion='gini', splitter='best', min_samples_split=10)
    clf = clf.fit(X, y)
    #clf = DecisionTreeClassifier().fit(X, y)

    # Plot the decision boundary
    plt.subplot(2, 3, pairidx + 1)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    print(x_min, x_max)
    print(y_min, y_max)
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)

    plt.xlabel(vec.feature_names_[pair[0]])
    plt.ylabel(vec.feature_names_[pair[1]])
    plt.axis("tight")

    # Plot the training points
    for i, color in zip(range(n_classes), plot_colors):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c=color, label=target_name[i],
                    cmap=plt.cm.Paired)

    plt.axis("tight")

plt.suptitle("Decision surface of a decision tree using paired features")
plt.legend()
plt.show()
```

# Using sklearn-pandas to transform the data

The idea here is that we use a wrapper that will take care of all transformations between pandas and numpy array. We can aply all the scikit-learn transformations to the pandas dataframe directly without explicitly converting it to numpy array.
This requires sklearn-pandas package:

https://github.com/paulgb/sklearn-pandas


```python
from sklearn_pandas import DataFrameMapper, cross_val_score
import sklearn.preprocessing
```

First, we will construct a mapper, which is a mapping between column names and transformation.

```python
mapper = DataFrameMapper([
...     ('c', sklearn.preprocessing.LabelBinarizer()),
...     (['a','b'], sklearn.preprocessing.StandardScaler())
... ])
```

Here is how the instance mapper looks like.

```python
mapper
```

Using method `fit_transform` to learn the transformation from the train dataset. You can think of `mapper` is currently a "learned model" and `train2` is the output of the transformation on the train dataset.

```python
train2 = mapper.fit_transform(df_train)
train2
```

Next, step we apply the learned transformation `mapper` to the test dataset simply using method `transform`.

```python
test2 = mapper.transform(df_test)
test2
```

Now, you `train2` and `test2` are in a proper format for scikit-learn, and we can build a model:

```python
clf2 = tree.DecisionTreeClassifier(criterion='gini', splitter='best', min_samples_split=20)
clf2 = clf2.fit(train2, y_train)
```

Nothing is new here, we just want to visualize the learned decision tree.

```python
## Extract the decision tree logic from the trained model
dot_data2 = StringIO() 
tree.export_graphviz(clf2, out_file=dot_data2) 

## convert the logics into graph
graph2 = pydot.graph_from_dot_data(dot_data2.getvalue()) 

## This will plot decision tree in pdf file
graph2.write_pdf(path="test_decistion_tree_2.pdf") 

from IPython.display import Image #needed to render in notebook
Image(graph2.create_png())
```

Although this latter model is learned from standardized data `train2`, the resulting tree is no different from the previous one (`train` without standardizing). That's because the split is done based on each variable alone, hence invariant to linear transformation.
