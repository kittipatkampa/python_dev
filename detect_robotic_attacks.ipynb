```python
## import the data into dataframe using pandas package
import pandas as pd
import numpy as np
```

```python
## import the input text file, format time, rename columns for convenience
din = pd.read_csv("myinput.txt", sep='\t')
din.TimeStamp = pd.to_datetime(din.TimeStamp)
din.columns = ['id','timestamp','count']

## sort by ID and timestamp
din = din.sort(columns=['id','timestamp'],ascending=[1,0])
```

```python
din
```

```python
## ==========================================================
## Calculate the accumalative count in any 10-min interval
## ==========================================================
din['intv_count'] = 0
min_intv = 10

for i in range(len(din.index)-1):
    ## initializing the values at current index i
    cur_id = din.id.iloc[i]
    intv_count = din['count'].iloc[i]
    j = i + 1
    
    while (j < len(din.index)) and (cur_id==din.id.iloc[j]):
        ## time difference in minutes
        min_diff = (din.timestamp.iloc[i]-din.timestamp.iloc[j]).total_seconds()/60
        if min_diff > min_intv:
            break
        else:
            intv_count = intv_count + din['count'].iloc[j]
        j += 1  
    ## store the count
    din.intv_count.iloc[i] = intv_count 

## display the result 
din    
```

```python
## print suspicious ID
positive_result = unique(din.id[din.intv_count >= 500])
print('list of robots detected:'+ str(positive_result))
```
