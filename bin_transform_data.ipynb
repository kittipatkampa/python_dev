# A better way to do quantile binning

It is not safe to use qcut on some data array. Below is an example where qcut works only for n < 5. That is because when n >= 5, the bin edges/separators are not unique, hence qcut gives error message:

ValueError: Bin edges must be unique: array([-99., -99.,  79.,  82.,  84.,  99.])

```python
n = 5
pd.qcut(X2,n)
```

Therefore, it's much safer to do the following.

```python
## first find the valid quantile edges, then dedup them
n = 10
bin_sep = X2.quantile(np.linspace(0.0, 1.0, num=n))
print('bin_sep w/o dedup\n',bin_sep)
bin_sep = bin_sep.unique()
print('bin_sep after deduped\n',bin_sep)

## Now you can use the bin_sep to factor the data
factor, bin_edges = pd.cut(x=X2, bins=bin_sep, right=True, include_lowest=True, retbins=True)
print('factor',factor)
print('bin_edges',bin_edges)
pd.DataFrame({'X2':X2, 'factor':factor})
```

# Calculate Weight of Evidence -- Spearman correlation

Start with a large number of intervals, keep reducing the number of interval until the Spearman correlation between X.mean and Y.mean is 1. 

Example here is adapted from 

https://statcompute.wordpress.com/2012/12/08/monotonic-binning-with-python/

```python
import scipy.stats.stats as stats
n = 100
r = 0
while np.abs(r) < 1:
    bin_sep = X2.quantile(np.linspace(0.0, 1.0, num=n))
    bin_sep = bin_sep.unique()
    factor,bins = pd.cut(x=X2, bins=bin_sep, right=True, include_lowest=True, retbins=True, precision=5)
    d1 = pd.DataFrame({"X": X2, "Y": Y, "Bucket": factor })
    d2 = d1.groupby('Bucket', as_index = True)
    r, p = stats.spearmanr(d2.mean().X, d2.mean().Y)
    print('n=',n,' r=',r)
    n = n-1

## Bucket, #neg, #pos, ratio_bad_ovr_good
counts = d2.size()
num_pos = d2.sum().Y
num_neg = d2.size()-num_pos
ratio_pos_ovr_neg = num_pos/num_neg
total_pos = sum(num_pos)
total_neg = sum(num_neg)
rel_p_pos = num_pos/total_pos
rel_p_neg = num_neg/total_neg
WoE = 100*np.log(rel_p_pos/rel_p_neg)

## report the results
print('final bins=',bins)
WoE_tbl = pd.DataFrame({'counts':counts, 'num_pos':num_pos, 'num_neg':num_neg, 'ratio_pos_ovr_neg':ratio_pos_ovr_neg, 'rel_p_pos':rel_p_pos, 'rel_p_neg':rel_p_neg, 'WoE':WoE})
pd.DataFrame(data=WoE_tbl, columns=['counts','num_pos','num_neg','ratio_pos_ovr_neg','rel_p_pos','rel_p_neg','WoE']) 
```

# Merging intervals using Chi-square test

$H_0$ -- A and B are independent P(A,B) = P(A)P(B)

$H_a$ -- A and B are not independent P(A,B) $\ne$ P(A)P(B)
      

If p-value >= 0.05 --> var1 and var2 are independent, hence f1 and f2 are from the same distribution.
If p-value < 0.05 --> var1 and var2 are NOT independent, hence f1 and f2 are from different distributions.
Note: f1 and f2 are treated as value instances for var2. Chi$^2$ test the independence of var1 and var2, not f1 and f2.

Example#1: The result shows p-value = 1.0, which is greater than 0.05. Therefore, we fail to reject the null hypothesis, hence two variables (var1 and var2) are independent and so f1 and f2 are said to come from the same distribution. 


```python
import scipy.stats as sp
f1 = np.array([80,20])
f2 = np.array([8,2])
conting = np.array([f1,f2])
##             var1
##            0    1
##      --------------
## var2 f1 | 80   20
##      f2 |  8    2
##      --------------
out = sp.chi2_contingency(conting)
out[0] ## chi2 test statistics
out[1] ## p-value of the test
out[2] ## Degrees of freedom
out[3] ## expected frequencies
print('chi2 statistics',out[0],' p-value',out[1],' degree of freedom',out[2],' expected freq',out[3])
```

Example#2 The p-value = 0.02 which is less than 0.05, therefore, we will reject the null hypothesis, two variables are not indepndent, hence f1 and f2 are not from the same distribution.

```python
f1 = np.array([80,20])
f2 = np.array([8,8])
conting = np.array([f1,f2])
out = sp.chi2_contingency(conting)
print('chi2 statistics',out[0],' p-value',out[1],' degree of freedom',out[2],' expected freq',out[3])
```

Example#3 It's more forgivable when having more data points. When using 800,200, the range within p-value=0.05 is wider than 80,20.

```python
p1 = 80
p2 = 20
epsilon_list = np.arange(-20,20,1)

def cal_p_val(p1,p2,epsilon_list):
    f1 = np.array([p1,p2])
    p_value = np.zeros_like(epsilon_list)*1.
    for i in np.arange(len(epsilon_list)):
        epsilon = epsilon_list[i]
        f2 = np.array([p1,p2+epsilon])
        #print(f2)
        conting = np.array([f1,f2])
        out = sp.chi2_contingency(conting)
        #print(out[1])
        p_value[i] = out[1]
        #print('i=',i,'  epsilon=',epsilon,'  p-value=',p_value[i])
    return (p_value)

epsilon_list1 = np.arange(-20,20,1)
p_value1 = cal_p_val(p1=80,p2=20,epsilon_list=epsilon_list1)
epsilon_list2 = np.arange(-200,200,1)
p_value2 = cal_p_val(p1=800,p2=200,epsilon_list=epsilon_list2)
plt.plot(epsilon_list1,p_value1, 'k-', label='80, 20')        
plt.plot(epsilon_list2,p_value2, 'r-', label='800,200')

```
