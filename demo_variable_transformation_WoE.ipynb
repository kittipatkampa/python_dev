### numerical variable transformation

```python
## =================================================================================
##
## functions for numerical variable transformation
##
## =================================================================================

def make_numer_profile(X, Y, n=20, p_val_thr=0.7, intvl_width_thr=2e-18, prior_num_pos=0.001, prior_num_neg=1, display_output=True, display_all_details=False):
    ''' 
    outputs:
    profile_tbl --  pandas.DataFrame; The final dataframe (after all the merging sequences) containing rich information 
                    about the variable X, for instance intervals, bin edges, min/max values in each bin, 
                    relative frequencies, weight of evidence (WoE) in each bin.
    WoE_numer_null -- scalar; The WoE value for missing-value (NULL/NAN).
    bin_sep -- pandas.Series; The final bin separators/edges corresponding to profile_tbl. This array presents the same 
                    information as bin_edge_left union bin_edge_right in profile_tbl.
    
    parameters:
    X -- pandas.Series, single variable from which you want to build WoE transformation table.
    Y -- pandas.Series, single target variable containing 0 or 1, corresponding to X
    n -- scalar, number of initial bins to start with
    p_val_thr -- scalar, the p-value threshold. Adjacent bins will be merge if the p-value (obtained from Chi^2 test) 
                 larger than p_val_thr. In other words, the larger p_val_thr is, the more bins we will have at the end. 
    intvl_width_thr -- scalar, the i+1 bin will be removed if its distance to i bin is less than the interval width threshold 
                       intvl_width_thr, whose default value is 2e-18. The value can be unstable when it is too small. It is safe 
                       to use larger threshold.
    prior_num_pos -- scalar, this is a loading/smoothening factor for calculating rel_p_pos in order to prevent -inf case when 
                    num_pos = 0. prior_num_pos should have relatively small value otherwise the calculation will be biased toward 
                    the smoothing parameter.
    prior_num_neg -- scalar, this is a loading/smoothening factor for calculating rel_p_neg in order to prevent inf case when 
                    num_neg = 0. prior_num_neg should have relatively small value otherwise the calculation will be biased toward 
                    the smoothing parameter.
    
    Note that prior_num_pos and prior_num_neg can be different values.  
    
    display_output -- boolean, if True, some short-messages will be displayed when running.
    display_all_details -- boolean, if True, the detail information will be displayed when running, which is good for debugging mode.
    
    *** This function requires package scipy.stats ***
    
    latest revised: Aug 21, 2014
    kittipat@             
    '''
    
    if display_output :
        print('\n')
        print('*'*99,'\n')
        print('summary on X and Y\n')
        print('X: size:',len(X),' #not-NULL:',sum(pd.notnull(X)),' #NULL:',sum(pd.isnull(X)))
        print('Y: size:',len(Y),' #NULL:',sum(pd.isnull(Y)),' #0:',sum(Y==0),' #1:',sum(Y==1))

    ## Get records that are not NULL
    X1 = X[pd.notnull(X)]
    Y1 = Y[pd.notnull(X)]

    if display_output :
        ## data sanity check
        print('\n')
        print('-'*99,'\n')
        print('processing not-null data\n')  
        print('X1 size:',len(X1),' #not-NULL:',sum(pd.notnull(X1)),' #NULL',sum(pd.isnull(X1)))
        print('Y1 size:',len(Y1),' #not-NULL:',sum(pd.notnull(Y1)),' #NULL',sum(pd.isnull(Y1)),' #0:',sum(Y1==0),' #1:',sum(Y1==1))

    ## number of positive, negative records
    total_counts = len(Y)
    total_pos = sum(Y)
    total_neg = total_counts-total_pos

    ## =======================================================================
    ## This part process only not-null records X1,Y1
    ## =======================================================================

    ## create bin edges
    ## if python3.4 use the one below
    ## -----------------------------------------------------
    ## [for python 3.4] -- this works well for python3.4, but not 2.7
    bin_sep = X1.quantile(np.linspace(0.0, 1.0, num=n)) 
    bin_sep = bin_sep.unique()
    ## if python2.7 use the one below
    ## -----------------------------------------------------
    ## [for python 2.7] -- So, I will need to use this version that works for python 2.7
    #bin_sep = np.percentile(X1.as_matrix(), np.linspace(0.0, 100.0, num=n))
    #bin_sep = np.unique(bin_sep)
    ## -----------------------------------------------------
    
    ## remove adjacent bins that are too close (closer than intvl_width_thr)
    intvl_width = np.abs(np.roll(bin_sep,-1) - bin_sep)
    if display_all_details : print('min interval difference: ',min(intvl_width))
    selected_idx = (intvl_width >= intvl_width_thr)
    selected_idx = np.roll(selected_idx,1)
    bin_sep = bin_sep[selected_idx]
    if display_all_details : print('initial bin_sep:\n',bin_sep)

    ## merge some bins that are zeros
    factor,bins = pd.cut(x=X1, bins=bin_sep, right=True, include_lowest=True, retbins=True, precision=5)
    bin_count = np.bincount(factor.labels, minlength=(len(bins)-1))
    idx_to_keep = np.append([True],bin_count > 0)
    bin_sep = bin_sep[idx_to_keep]
    
    ## while loop to merge similar adjacent bins
    done = 0 ## indicator
    while done == 0:
        factor,bins = pd.cut(x=X1, bins=bin_sep, right=True, include_lowest=True, retbins=True, precision=5)
        d1 = pd.DataFrame({"X": X1, "Y": Y1, "Bucket": factor })
        d2 = d1.groupby('Bucket', as_index = True)

        ## Bucket, #neg, #pos, ratio_bad_ovr_good
        counts = d2.size()
        num_pos = d2.sum().Y
        num_neg = counts-num_pos
        ratio_pos_ovr_neg = num_pos/num_neg
        
        ## calculate smoothened relative frequency of positive and negative class
        rel_p_pos = (num_pos + prior_num_pos)/(total_pos + prior_num_pos*(len(bins)-1) )
        rel_p_neg = (num_neg + prior_num_neg)/(total_neg + prior_num_neg*(len(bins)-1) )
        WoE = np.log(rel_p_pos/rel_p_neg)
        
        bin_edge_left = bins[:(len(bins)-1)]
        bin_edge_right = bins[1:len(bins)]
        min_intv = d2.min().X
        max_intv = d2.max().X

        ## output the results
        #print('final bins=',bins)
        profile_dict = pd.DataFrame({'counts':counts, 'num_pos':num_pos, 'num_neg':num_neg, 'ratio_pos_ovr_neg':ratio_pos_ovr_neg, 'rel_p_pos':rel_p_pos, 'rel_p_neg':rel_p_neg, 'WoE':WoE, 'min_intv':min_intv,'max_intv':max_intv})
        profile_tbl = pd.DataFrame(data=profile_dict, columns=['min_intv','max_intv','counts','num_pos','num_neg','ratio_pos_ovr_neg','rel_p_pos','rel_p_neg','WoE']) 
        ## sort the interval by min_value
        profile_tbl = profile_tbl.sort_index(by=['min_intv'])
        ## After sort, we can add the bin edge
        profile_tbl['bin_edge_left'] = bin_edge_left
        profile_tbl['bin_edge_right'] = bin_edge_right

        ## add auxiliary columns i, j and p-value
        profile_tbl['i'] = np.zeros(len(profile_tbl))
        profile_tbl['j'] = np.zeros(len(profile_tbl))
        profile_tbl['adjc_p_value'] = np.zeros(len(profile_tbl))

        ## ch^2-test and calculate p-value for each adjacent row pair
        i = 0
        while i < (len(profile_tbl)-1):
            #print(i)
            conting = profile_tbl[['num_pos','num_neg']].iloc[[i,i+1]]
            ## put condition to prevent chi^2 error due to [0,x] vs [0,y]
            if (conting.num_pos.sum()==0) | (conting.num_neg.sum()==0):
                profile_tbl.adjc_p_value[i] = 1
            else:
                profile_tbl.adjc_p_value[i] = sp.chi2_contingency(conting)[1]
            
            profile_tbl.i[i] = i
            profile_tbl.j[i] = i+1
            #print(i,',',i+1,' p-value:',profile_tbl.adjc_p_value[i])
            i = i + 1

        if display_all_details : print('intermediate profile_tbl:\n',profile_tbl)    
        ## Get the max p-value with corresponding index
        idx_p_val_max = profile_tbl.adjc_p_value.idxmax()
        p_val_max = profile_tbl.adjc_p_value[idx_p_val_max] ## the p-value max
        idx_to_remove = profile_tbl.j[idx_p_val_max] ## the index we want to remove in bins

        ## Compare the max p-value with the threshold
        if p_val_max >= p_val_thr:
            ## merge the interval i and j
            ## This is a new bin separators
            bin_sep = np.delete(bins,idx_to_remove)
            if display_all_details : print('p_val_max=',p_val_max,'>= p_val_thr=',p_val_thr,'  continue merging intervales')
            done = 0
        else:
            ## all intervals are sufficiently unique, no more mergeing, done!!!  
            if display_all_details : print('p_val_max=',p_val_max,'< p_val_thr=',p_val_thr,'  done merging intervales!!!')
            done = 1


        if display_all_details : print('current bin separators',bins)
        if display_all_details : print('new bin separators',bin_sep)

    if display_output :        
        ## final outputs for not-null records
        #print('\n')
        #print('-'*99,'\n')
        print('\nfinal bin for not-null data: \n',bin_sep)
        print('\nfinal profile table for not-null data: \n',profile_tbl)  
        #print('\n')
        #print('-'*99,'\n')
    
    
    
    ## =======================================================================
    ## This part process only null records X2,Y2
    ## =======================================================================

    ## Get null records
    X2 = X[pd.isnull(X)]
    Y2 = Y[pd.isnull(X)]

    if display_output :
        ## data sanity check
        print('\n')
        print('-'*99,'\n')
        print('processing null data\n')  
        print('X2 size:',len(X2),' #not-NULL:',sum(pd.notnull(X2)),' #NULL',sum(pd.isnull(X2)))
        print('Y2 size:',len(Y2),' #not-NULL:',sum(pd.notnull(Y2)),' #NULL',sum(pd.isnull(Y2)),' #0:',sum(Y2==0),' #1:',sum(Y2==1))


    ## calculate necessary metrics for null data
    rel_p_pos = sum(Y2==1)/total_pos
    rel_p_neg = sum(Y2==0)/total_neg
    WoE_numer_null = np.log(rel_p_pos/rel_p_neg)

    ## if there is no null value, assign WoE = 0
    if pd.isnull(WoE_numer_null): WoE_numer_null = 0
    
    if display_output :
        #print(rel_p_pos, total_pos)
        #print(rel_p_neg, total_neg)
        #print('\n')
        #print('-'*99,'\n')
        print('\nWoE for null data:',WoE_numer_null)
        print('='*99,'\n')
    
    ## return the output
    return (profile_tbl, WoE_numer_null, bin_sep)


def X_to_wX_numer(X_new, bin_sep, WoE, WoE_numer_null, WoE_outbound_low='lowest', WoE_outbound_high='highest', display_output=True, display_all_details=False):
    '''
    outputs:
    wX -- pandas.Series; the variable X_new after transformation into weight of evidence space.
    bin_sep_extended -- numerical array; the final bin used to bin the variable X_new
    
    parameters:
    X_new -- pandas.Series; the variable we want to convert to WoE
    bin_sep -- the input bin separators/edges we want to use to bin the input X_new
    WoE -- pandas.Series; The WoE corresponding to bin_sep
    WoE_numer_null -- WoE for null value
    WoE_outbound_low='lowest' -- WoE for the case that the input value is less than min(bin_sep)
    WoE_outbound_high='highest' -- WoE for the case that the input value is greater than max(bin_sep) 
    display_output -- boolean, if True, some short-messages will be displayed when running.
    display_all_details -- boolean, if True, the detail information will be displayed when running, which is good for debugging mode.
    
    Note: the index of output wX would remain the same as its input X_new 
    
    latest revised: Aug 21, 2014
    kittipat@
    '''
    
    ## =======================================================================
    ## This part converts X_new to wX_new
    ## =======================================================================
    
    WoE_outbound_low = WoE[0] # 0 or min(WoE) is fine
    WoE_outbound_high = WoE[len(WoE)-1] # 0 or max(WoE) is fine

    ## extend the bin_sep to -Inf and +Inf
    if display_all_details : print('\nbin_sep: \n',bin_sep)
    bin_sep_extended = np.append(-np.inf,bin_sep) 
    bin_sep_extended = np.append(bin_sep_extended, np.inf)
    if display_output : print('\nbin_sep_extended: \n',bin_sep_extended)

    ## prepare the output wX
    wX = pd.Series(np.zeros(len(X_new))*1., index=X_new.index)
    if display_all_details : print('wX index:',wX.index)
    
    ## ------------------------------------------------
    ## part 1 -- transform missing value
    ## ------------------------------------------------
    idx_null = pd.isnull(X_new)
    if display_all_details : print('idx_null index:',idx_null.index)
    wX[idx_null] = WoE_numer_null
    

    
    ## ------------------------------------------------
    ## part 2 -- transform not-null value
    ## ------------------------------------------------
    ## for loop for each interval
    if display_all_details : print('\nlen(bin_sep_extended)-1:',len(bin_sep_extended)-1)
    for i in range(len(bin_sep_extended)-1):
        a = bin_sep_extended[i]
        b = bin_sep_extended[i+1]
        idx_fill = (a < X_new) & (X_new <= b)
        
        if i == 0:
            ## [-Inf,first]
            if display_all_details : print('i:',i,' ',a,',',b,' WoE:',WoE_outbound_low)
            wX[ idx_fill ] = WoE_outbound_low  
        elif i == (len(bin_sep_extended)-2):
            ## (last,Inf]
            if display_all_details : print('i:',i,' ',a,',',b,' WoE:',WoE_outbound_high)
            wX[ idx_fill ] = WoE_outbound_high
        else: ## (a,b]
            if display_all_details : print('i:',i,' ',a,',',b,' WoE:',WoE[i-1])
            wX[ idx_fill ] = WoE[i-1]

    ## return the outputs        
    if display_all_details : print(pd.DataFrame({'X_new':X_new,'factor':pd.cut(X_new,bin_sep_extended),'wX':wX}) )
    if display_all_details : print('wX length:',len(wX))
    return (wX,bin_sep_extended)

```

### Categorical variable transformation

```python
## =================================================================================
##
## functions for categorical variable transformation
##
## =================================================================================

def make_categ_profile(X, Y, prior_num_pos=0.001, prior_num_neg=1, display_output=False, display_all_details=False):
    ''' 
    outputs:
    profile_tbl --  pandas.DataFrame; The final dataframe containing rich information about the variable X, 
                    for instance intervals, counts of positive/negative samples (num_pos,num_neg) in each bin, 
                    relative frequencies, weight of evidence (WoE) in each bin.
    
    parameters:
    X -- pandas.Series, single variable from which you want to build WoE transformation table.
    Y -- pandas.Series, single target variable containing 0 or 1, corresponding to X.
    prior_num_pos -- scalar, this is a loading/smoothening factor for calculating rel_p_pos in order to prevent -inf case when 
                    num_pos = 0. prior_num_pos should have relatively small value otherwise the calculation will be biased toward 
                    the smoothing parameter.
    prior_num_neg -- scalar, this is a loading/smoothening factor for calculating rel_p_neg in order to prevent inf case when 
                    num_neg = 0. prior_num_neg should have relatively small value otherwise the calculation will be biased toward 
                    the smoothing parameter.
    
    Note: 
    1) prior_num_pos and prior_num_neg can be different values. 
    2) Since this code uses groupby to bin the data and it ignores the missing data (Null, NaN), therefore, the code convert the
        missing data to the string 'Null_char_python'. You can also see that the same string will be used with identical purpose 
        when we transform value to wX as well.
    
    latest revised: Aug 21, 2014
    kittipat@             
    '''

    ## number of positive, negative records
    total_counts = len(Y)
    total_pos = sum(Y)
    total_neg = total_counts-total_pos

    ## counting each category
    X = X.fillna('Null_char_python') 
    d1 = pd.DataFrame({'X':X,'Y':Y})
    d2 = d1.groupby('X', as_index = True)

    ## Bucket, #neg, #pos, ratio_bad_ovr_good
    counts = d2.size()
    num_pos = d2.sum().Y
    num_neg = counts-num_pos
    ratio_pos_ovr_neg = num_pos/num_neg

    ## calculate smoothened relative frequency of positive and negative class
    rel_p_pos = (num_pos + prior_num_pos)/(total_pos + prior_num_pos*(counts) )
    rel_p_neg = (num_neg + prior_num_neg)/(total_neg + prior_num_neg*(counts) )
    WoE = np.log(rel_p_pos/rel_p_neg)

    ## output the results
    profile_dict = pd.DataFrame({'counts':counts, 'num_pos':num_pos, 'num_neg':num_neg, 'ratio_pos_ovr_neg':ratio_pos_ovr_neg, 'rel_p_pos':rel_p_pos, 'rel_p_neg':rel_p_neg, 'WoE':WoE})
    profile_tbl = pd.DataFrame(data=profile_dict, columns=['counts','num_pos','num_neg','ratio_pos_ovr_neg','rel_p_pos','rel_p_neg','WoE']) 

    return profile_tbl


def X_to_wX_categ(X_new, WoE_categ, WoE_categ_unseen = 0, display_output=False, display_all_details=False):
    '''
    outputs:
    wX -- pandas.Series; the variable X_new after transformation into weight of evidence space.
    
    parameters:
    X_new -- pandas.Series; the str (type = 'object') variable we want to convert to WoE
    WoE_categ -- pandas.Series or dict(); The WoE corresponding to each bin
    WoE_categ_unseen -- scalar, the WoE value for unseen categorical value. Don't confuse this with NaN or Null.
    
    Note: the index of output wX would remain the same as its input X_new 
    
    latest revised: Aug 21, 2014
    kittipat@
    '''
    
    ## it's required that WoE_categ is a dict
    if not( isinstance(WoE_categ, type(dict())) ):
        WoE_categ = WoE_categ.to_dict()
    
    X_new = X_new.fillna('Null_char_python') ## Null value mapped to 'Null_char_python'
    wX = X_new.map(WoE_categ)
    wX[wX.isnull()] = WoE_categ_unseen ## transform unseen values
    
    return wX
```

```python
## [user-defined] the filename of the output WoE table
report_filename = 'titanic_WoE_table.txt'
## [user-defined] List of the variable names you want to transform
## we want all the variables
variable_name_list = df.columns 
## we want subset of variables
#variable_name_list = ['pclass', 'sex', 'age', 'sibsp', 'parch', 'ticket', 'fare', 'cabin', 'embarked']
```

### Loop to create WoE transformation table and transform variables

```python
## [production code]
## ------------------------------------------------------ 
## The code below shows how to make WoE transformation table and transform the data into WoE space. 
## Converting each column into WoE using for-loop
## ------------------------------------------------------


## List variable names that will be transformed to WoE
var_all = df_train.columns ## list of all variables in the dataframe

## prepare the empty dataframe for output
wX_valida = pd.DataFrame(data=None, index=df_valida.index, columns=variable_name_list)
wX_train = pd.DataFrame(data=None, index=df_train.index, columns=variable_name_list)

## open file connection here
out_file = open(report_filename, 'wt')
out_file.write('\n'+'-'*40 + report_filename +'-'*40 + '\n')
out_file.close() 

## prepare empty dict for profile
profile_tbl_dict = {}
cnt = 1

for v in variable_name_list:
    
    ## variable name
    variable_name = v
    
    ## open file connection to append the file
    out_file = open(report_filename, 'a')
    out_file.write('\n'+'-'*99 + '\n')
    
    ## check if the variable exists in the dataframe, if not, then skip to the next one
    if not(variable_name in var_all): 
        ## display status
        print('\nprocessing',str(cnt),'/',str(len(variable_name_list)),variable_name,'...')
        out_file.write(str(cnt)+'/'+str(len(variable_name_list))+' '+variable_name+' is not in the dataframe df_train hence will be skipped, and the column will be excluded from the output'+ '\n')
        wX_train.drop(variable_name, axis=1, inplace=True)
        wX_valida.drop(variable_name, axis=1, inplace=True)
        cnt = cnt + 1 ## update counter
        out_file.close()
        continue ## skip this variable
    
    ## Get X and Y and X_new
    X = df_train[variable_name] 
    X_new = df_valida[variable_name]


    ## display status
    out_file.write('\nprocessing '+str(cnt)+'/'+str(len(variable_name_list))+' '+variable_name+' '+str(X.dtypes)+' ...' + '\n')
    print('\nprocessing',str(cnt),'/',str(len(variable_name_list)),variable_name,X.dtypes,'...')
    
    ## check the type of Series (numerical vs categorical)
    
    if ( X.dtypes=='float64' ) | (( X.dtypes=='int64' ) & (len(X.unique()) >= 30) ) :
        ## -------------------------------
        ## ----- for numerical value ----- 
        ## -------------------------------
        ## Make WoE transformation table
        profile_tbl, WoE_numer_null, bin_sep = make_numer_profile(X, Y, n=20, p_val_thr=0.7, intvl_width_thr=1e-9, display_output=False)
        out_file.write('\nfinal bin separators:\n' + str(bin_sep) + '\n')
        
        ## store the WoE transformation table in a dict
        profile_tbl_dict[variable_name] = profile_tbl
        profile_tbl_dict[variable_name+'_null'] = WoE_numer_null

        ## Convert X to wX and X_new to wX_new
        WoE = profile_tbl.WoE
        out_file.write('\nfinal WoE profile:\n'+ str(profile_tbl[['num_pos','num_neg','rel_p_pos','rel_p_neg','WoE']]) + '\n')
    
        out_file.write('\ntransforming train dataset...\n')
        wX_train_tmp, bin_sep_out = X_to_wX_numer(X, bin_sep, WoE, WoE_numer_null, display_output=False, display_all_details=False)
    
        out_file.write('\ntransforming validation dataset...\n')
        wX_valida_tmp, bin_sep_out = X_to_wX_numer(X_new, bin_sep, WoE, WoE_numer_null, display_output=False, display_all_details=False)
 
    elif (X.dtype == 'object') | ( ( X.dtypes=='int64' ) & (len(X.unique()) < 30) ) :
        ## -------------------------------
        ## ----- for categorical value ----- 
        ## -------------------------------
        ## Make WoE transformation table for categorical
        profile_tbl_categ = make_categ_profile(X, Y, prior_num_pos=0.001, prior_num_neg=1)
        
        ## store the WoE transformation table in a dict
        profile_tbl_dict[variable_name] = profile_tbl_categ
        
        ## Convert X to wX and X_new to wX_new
        WoE_categ = profile_tbl_categ.WoE.to_dict()
        out_file.write('\nfinal WoE-categorical profile:\n'+str(profile_tbl_categ[['num_pos','num_neg','rel_p_pos','rel_p_neg','WoE']])+'\n')
        
        out_file.write('\ntransforming train dataset...\n')
        wX_train_tmp = X_to_wX_categ(X, WoE_categ, WoE_categ_unseen = 0)
        
        out_file.write('\ntransforming validation dataset...\n')
        wX_valida_tmp = X_to_wX_categ(X_new, WoE_categ, WoE_categ_unseen = 0)
    
    else: 
        out_file.write('the object of type '+str(X.dtypes)+' is not available for this function, send email to kittipat@gmail.com\n')  
    
    ## prepare the output dataframe
    wX_train[v] = wX_train_tmp
    wX_valida[v] = wX_valida_tmp
    out_file.write('\n'+'='*99 + '\n')
    
    ## close the file connection
    out_file.close()
    
    ## update counter
    cnt = cnt + 1

    
```
